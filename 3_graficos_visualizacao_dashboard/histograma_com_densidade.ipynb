# Gráfico de Densidade

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os dados
base = pd.read_csv('Dados/trees.csv')
base.head()
```

```python
# Gerando um Histograma do Height
plt.hist(base.iloc[:,1], bins = 6)
```

```python
# Vou fazer o mesmo agora como seabor
#kde = linha de edensidade
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title='Árvores')
```

```python
# Agora somente com a linha
sns.kdeplot(base.iloc[:,1], bins=6, color='blue').set(title='Árvores')
```

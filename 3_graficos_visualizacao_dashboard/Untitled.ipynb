# Histograma com densidade

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('Dados/chicken.csv')
base
```

```python
# Criando um dataframe agrupando o atributto feed
agrupado = base.groupby(['feed'])['weight'].sum()
agrupado
```

```python
# Histograma considerando somente um alimento
srn.kdeplot(base.loc[base['feed'] == 'horsebean'].weight).set(title='horsebean')
```

```python
srn.kdeplot(base.loc[base['feed'] == 'casein'].weight).set(title='casein')
```

```python
srn.kdeplot(base.loc[base['feed'] == 'linseed'].weight).set(title='linseed')
```

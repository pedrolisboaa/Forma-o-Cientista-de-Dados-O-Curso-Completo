# Gráfico de Dispersão

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dado = pd.read_csv('Dados/trees.csv')
dado.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=dado.Girth, y=dado.Volume, color='grey', facecolor='none', marker= '*')
plt.title('Árvore')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
# Sem o mark ele fica diferentte
plt.scatter(x=dado.Girth, y=dado.Volume, color='grey', facecolor='none')
plt.title('Árvore')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Agora o mesmo com seaborn
sns.regplot(x=dado.Girth, y=dado.Volume, data=dado, x_jitter=0.3, fit_reg=False)
```
